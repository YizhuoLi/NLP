### Now begin to get the subway data

```python
# get the metro url first
import requests

headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.103 Safari/537.36'
}

base_url = 'https://baike.baidu.com/item/%E5%8C%97%E4%BA%AC%E5%9C%B0%E9%93%81/408485#2'

base_req = requests.get(base_url, headers = headers)
```

```python
base_req.encoding='utf-8'
# base_req.text
```

```python
# here we assume that every subway page has the detail.
base_text = base_req.text.replace('\n','')
base_text_refine = re.findall('一个动车和一个拖车为一个制动单元.*?</table>', base_text)[0]
base_text_refine
```

```python
from scrapy.selector import Selector

Selector(text=base_text_refine).xpath('//tr/td/a[@href]').extract()
```

```python
base_list = Selector(text=base_text_refine).xpath('//tr/td/a[@href]').extract()
error_list = []

error_num_connection = collections.defaultdict(list)

stations_connection = collections.defaultdict(list)
dist = collections.defaultdict(int)
node_in_line = collections.defaultdict(set)

for base in base_list:
    line = re.findall('>(.*)<', base)[0]
    url = re.findall('href="(.*)"', base)[0]
    url = 'https://baike.baidu.com'+str(url)
    
    req = requests.get(url, headers = headers)
    req.encoding='utf-8'
    text = req.text.replace('\n','')
    
    try:
        text_refine = re.findall('相邻站间距信息统计表.*?</table>', text)[0]

        stations = Selector(text=text_refine).xpath('//tr/th/text()').extract()[3:]
        distances = Selector(text=text_refine).xpath('//tr/td[1]/text()').extract()
    #     print(stations)
    #     print(distances)
    #         directors = Selector(text=text_refine).xpath('//tr/td[2]/text()').extract()

        if len(stations) == len(distances):
            print('\n----The information of {} is following...'.format(line))
            for i in range(len(stations)):
    #             start, end = re.findall(re.compile('(\w+).*?(\w+)'), stations[i])[0]
                try:
                    start, end = stations[i].split('——')      

                    stations_connection[start].append(end)
                    stations_connection[end].append(start)

                    distance = distances[i].replace('米','')
                    if '千' in distance:
                        distance = float(distance.replace('千',''))*1000

                    dist[(start,end)] = dist[(end,start)] = int(distance)

                    node_in_line[start].add(line)
                    node_in_line[end].add(line)
                    print('{}--{}: {}'.format(start, end, distances[i]))
                except:
                    error_num_connection[line].append(i)
                    print('\n $$$$$$$$$$ ', line, 'num ',i, ' is wrong')


        else:
            error_list.append(line)
            print('\n', line, 'length not the same')        
        
    except:
        error_list.append(line)
        print('\n', line, ' without 相邻站间距信息统计表')
```
